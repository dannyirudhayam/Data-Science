```python
# 1.

import pandas as pd

file_path = "C:/Users/XNovaQ.24/Downloads/students_assignment_data.xlsx"
df = pd.read_excel(file_path)

print(df)
```

```python
# 2.

print(df.head(15))

print(df.dtypes)

print(df.shape)
```

```python
# 3.

print("\n Unique Cities")
print(df['City'].unique())

print("\n Students per City")
print(df['City'].value_counts())
```

```python
# 4. 

high_scorers = df[df['Total_Marks'] >= 400]
print("\n Students with Total_Marks >= 400 \n")
print(high_scorers)
```

```python
# 5. 

female_students = df[df['Gender'] == 'Female']
print("\n Female Students")
print(female_students)
```

```python
# 6.

avg_gender = df.groupby('Gender')['Total_Marks'].mean()
print("\n Average Marks by Gender")
print(avg_gender)
```

```python
# 7.

city_group = df.groupby('City').agg(
    Avg_Marks=('Total_Marks', 'mean'),
    Count=('StudentID', 'count')
)

print("\n City-wise Avg Marks & Count \n")
print(city_group)
```

```python
# 8. 

sorted_df = df.sort_values(by='Total_Marks', ascending=False).reset_index(drop=True)

print("\n Sorted by Total_Marks (Top 10) \n")
print(sorted_df.head(10))
```

```python
# 9. 
!pip install numpy
import numpy as np

df_missing = df.copy()
num_missing = int(0.10 * len(df_missing))

np.random.seed(42)   # Keep it reproducible
missing_indices = np.random.choice(df_missing.index, num_missing, replace=False)
df_missing.loc[missing_indices, 'Total_Marks'] = np.nan

print("\n Missing Values Introduced ")
print(df_missing['Total_Marks'].isna().sum())
```

```python
# 10. 

mean_marks = df_missing['Total_Marks'].mean()
df_missing['Total_Marks'] = df_missing['Total_Marks'].fillna(mean_marks)

print("\n Missing Values After Imputation ")
print(df_missing['Total_Marks'].isna().sum())
```

```python
# 11.

df_missing['Result'] = df_missing['Total_Marks'].apply(
    lambda x: 'Pass' if x >= 200 else 'Fail'
)

print("\n Result Column Added \n")
print(df_missing[['StudentID', 'Total_Marks', 'Result']].head())
```

```python
# 12.

print("\n Pass/Fail Count \n")
print(df_missing['Result'].value_counts())
```

```python
# 13.

df_missing.to_csv ("students_cleaned.csv", index=False)

print("\nFile saved as: students_cleaned.csv")
```
