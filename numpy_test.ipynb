```python
# 3.

arr10 = np.arange(100).reshape(10, 10)

rows = np.random.choice(10, 5)
cols = np.random.choice(10, 5)
adv_selected = arr10[rows, cols]

print("\nAdvanced Indexed Elements:\n", adv_selected)
```

```python
# 4. 

arr100 = np.arange(100)
arr10x10 = arr100.reshape(10, 10)
flat_again = arr10x10.flatten()

print("\nIntegrity Check:", np.array_equal(arr100, flat_again))
```

```python
a1 = np.ones((3, 3))
a2 = np.zeros((3, 3))
a3 = np.full((3, 3), 5)

concat_axis0 = np.concatenate((a1, a2, a3), axis=0)  # vertical
concat_axis1 = np.concatenate((a1, a2, a3), axis=1)  # horizontal

print("\nConcatenated Axis 0:\n", concat_axis0)
print("\nConcatenated Axis 1:\n", concat_axis1)

# Horizontal & Vertical Stack
print("\nHorizontal Stack:\n", np.hstack((a1, a2)))
print("\nVertical Stack:\n", np.vstack((a1, a2)))
```

```python
mat12 = np.arange(144).reshape(12, 12)
split_vertical = np.split(mat12, 3, axis=1)

print("\nVertical Splits (12x12 → Three 12x4):")
for i, part in enumerate(split_vertical):
    print(f"\nPart {i+1}:\n", part)
```

```python
# 5. 

arr_15x10 = np.random.randint(1, 100, (15, 10))

mean_cols = np.mean(arr_15x10, axis=0)
median_cols = np.median(arr_15x10, axis=0)
std_cols = np.std(arr_15x10, axis=0)

print("\nColumn Means:\n", mean_cols)
print("\nColumn Medians:\n", median_cols)
print("\nColumn Std:\n", std_cols)
```

```python
min_idx = np.argmin(arr_15x10, axis=1)
max_idx = np.argmax(arr_15x10, axis=1)

print("\nMin Indices per Row:\n", min_idx)
print("\nMax Indices per Row:\n", max_idx)
```

```python
flat = arr_15x10.flatten()
print("\nCumulative Sum:\n", np.cumsum(flat))
print("\nCumulative Product:\n", np.cumprod(flat))
```

```python
# 6.

arr = np.random.random(200)
nan_count = int(0.10 * len(arr))
nan_indices = np.random.choice(len(arr), nan_count, replace=False)
arr[nan_indices] = np.nan

print("\nTotal Missing Values:", np.isnan(arr).sum())
```

```python
mean_value = np.nanmean(arr)
arr_filled = np.where(np.isnan(arr), mean_value, arr)

print("\nMissing After Filling:", np.isnan(arr_filled).sum())
```

```python
arr_no_nans = arr[~np.isnan(arr)]
print("\nArray After Removing NaNs (size):", arr_no_nans.size)
```
